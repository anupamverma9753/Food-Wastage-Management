```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Loading Data

```python
# Load data

providers = pd.read_csv('/Users/anupamverma/Downloads/providers_data.csv')
receivers = pd.read_csv('/Users/anupamverma/Downloads/receivers_data.csv')
food_listings = pd.read_csv('/Users/anupamverma/Downloads/food_listings_data.csv')
claims = pd.read_csv('/Users/anupamverma/Downloads/claims_data.csv')
```

### Data Cleaning

```python
# Data Cleaning
# 1. Clean up newline characters in Address
providers['Address'] = providers['Address'].str.replace('\n', ', ', regex=False)

# 2. Extract extension from contact if present
providers['P_Extension'] = providers['Contact'].str.extract(r'x(\d+)')
# Strip extensions and special characters from contact numbers to clean them
providers['P_Clean_Contact'] = providers['Contact'].str.replace(r'x\d+', '', regex=True).str.replace(r'[\s\-\(\)\.]', '', regex=True)

providers.head()
```

```python
# Data Cleaning
# Extract extension from contact if present
receivers['R_Extension'] = receivers['Contact'].str.extract(r'x(\d+)')
# Strip extensions and special characters from contact numbers to clean them
receivers['R_Clean_Contact'] = receivers['Contact'].str.replace(r'x\d+', '', regex=True).str.replace(r'[\s\-\(\)\.]', '', regex=True)
```

### Preparing for SQL Export

```python
# Clean up categorical string columns across all tables for trailing Whitespaces
claims['Status'] = claims['Status'].str.strip()
food_listings['Food_Name'] = food_listings['Food_Name'].str.strip()
food_listings['Food_Type'] = food_listings['Food_Type'].str.strip()
food_listings['Meal_Type'] = food_listings['Meal_Type'].str.strip()
food_listings['Provider_Type'] = food_listings['Provider_Type'].str.strip()
providers['Type'] = providers['Type'].str.strip()
receivers['Type'] = receivers['Type'].str.strip()
```

```python
# Drop Provider type because when we JOIN it in SQL with Provider ID No need for the Provider_type Column
food_listings = food_listings.drop(columns=['Provider_Type'])
```

```python
food_listings
```

```python
receivers
```

```python
# Keep only the cleaned columns and rename them to final database names
providers = providers.drop(columns=['Contact_Number'])
providers = providers.rename(columns={'P_Clean_Contact': 'Contact_Number', 'P_Extension': 'Phone_Extension'})

receivers = receivers.drop(columns=['Contact_Number'])
receivers = receivers.rename(columns={'R_Clean_Contact': 'Contact_Number', 'R_Extension': 'Phone_Extension'})
```

```python
from sqlalchemy import create_engine

# Define your PostgreSQL connection string
DATABASE_URL = "postgresql://postgres:@localhost:5432/food_wastage"

# Create the engine
engine = create_engine(DATABASE_URL)

# Export dataframes to PostgreSQL tables
providers.to_sql('providers', engine, if_exists='replace', index=False)
receivers.to_sql('receivers', engine, if_exists='replace', index=False)
food_listings.to_sql('food_listings', engine, if_exists='replace', index=False)
claims.to_sql('claims', engine, if_exists='replace', index=False)

print("All four tables have been successfully exported to PostgreSQL!")
```
